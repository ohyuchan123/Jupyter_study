```python
import pandas as pd
url = 'https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84'
tables = pd.read_html(url, header=0, index_col=0) # header와 index column을 지정함
df_medals = tables[0]
df_medals
```

```python
summer = df_medals[["하계 참가 횟수", "Unnamed: 2","Unnamed: 3", "Unnamed: 4" ]]
summer
```

```python
summer.columns=['참가수','금','은','동']
summer
```

```python
summer["메달 총계"]=summer["금"]+summer["은"]+summer["동"]
summer
```

```python
summer.sort_values(["금","은","동"],ascending=False)
```

```python
summer["금비중"]=summer["금"]/summer["메달 총계"]
summer
```
