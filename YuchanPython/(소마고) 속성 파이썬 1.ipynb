**<연습문제>**

정수 하나를 변수 x에 입력받아, x가 11과 20 사이면 **‘11 ~ 20’**, 21과 30 사이면 **‘21~30’**, 아무것도 해당하지 않으면 **‘아무것도 해당하지 않음’** 이 출력되게 만드세요!

`````python
x=int(input("정수를 입력하세요>"))
if x>10 and x<=20:
    print('11~20')
elif x>20 and x<=30:
    print('21~30')
else:
    print('아무것도 해당하지 않음')
`````
