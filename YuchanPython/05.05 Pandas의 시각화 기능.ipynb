# 5.5 Pandas의 시각화 기능

Pandas의 시리즈나 데이터프레임은 `plot`이라는 시각화 메서드를 내장하고 있다. `plot`은 matplotlib를 내부에서 임포트하여 사용한다. 

```python
# sns, plt, mpl import
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 폰트 설정
mpl.rc('font', family='NanumGothic')
# 유니코드에서  음수 부호설정
mpl.rc('axes', unicode_minus=False)
```

```python
import numpy as np
import pandas as pd
np.random.seed(0)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2020', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```python
df1
```

```python
## pandas 데이터를 바로 plot할 수 있음!!!
df1.plot()
plt.title("Pandas의 Plot메소드 사용 예")
plt.xlabel("시간")
plt.ylabel("Data")
plt.show()
```

`plot` 메서드의 `kind`라는 인수를 바꾸면 여러가지 플롯을 그릴 수 있다. 다음과 같은 인수값을 지원한다.

* `bar`
* `pie`
* `hist`
* `kde`
* `box`
* `scatter`
* `area`


```python
# iris, titanic 데이터 불러옴
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터

iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
## iris pandas 데이터의 특정 열을 슬라이스하여 Series형태에서 바로 bar plot
```
