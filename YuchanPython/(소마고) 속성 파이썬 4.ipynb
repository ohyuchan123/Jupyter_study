#### (5) for 반복문을 사용하여 2차원 리스트 만들기

`````python
a = []   # 빈 리스트 생성

for i in range(3):
    line = []
    for j in range(2):
        line.append(0)
    a.append(line)
    
print(a)
`````

### 17.3 문자열 메서드

(1) `replace(바꿀 문자열, 새 문자열)`: 문자열 안의 문자열을 다른 문자열로 바꿈

(5) `lower()` : 소문자로
